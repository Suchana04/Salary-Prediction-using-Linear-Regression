```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sal_data = pd.read_csv("Dataset.csv")
sal_data.head()
```

```python
sal_data.shape
```

```python
sal_data.columns
```

```python
sal_data.columns = ['Age','Gender','Degree','Job_Title','Experience_years','Salary']
```

```python
sal_data.head()
```

```python
sal_data.dtypes
```

```python
sal_data[sal_data.duplicated()].shape
```

```python
sal_data1 = sal_data.drop_duplicates(keep = 'first')
```

```python
sal_data1.isnull().sum()
```

```python
sal_data1.dropna(how='any',inplace=True)
```

```python
sal_data1.shape
```

```python
sal_data1.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder = LabelEncoder()
```

```python
sal_data1['Gender_Encoded'] = Label_Encoder.fit_transform(sal_data1['Gender'])

sal_data1['Degree_Encoded'] = Label_Encoder.fit_transform(sal_data1["Degree"])

sal_data1['Job_Title_Encoded'] = Label_Encoder.fit_transform(sal_data1['Job_Title'])
```

```python
sal_data1.head()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
sal_data1['Age_scaled'] = std_scaler.fit_transform(sal_data1[['Age']])
sal_data1['Experience_years_scaled'] = std_scaler.fit_transform(sal_data1[['Experience_years']])
```

```python
sal_data1.head()
```

```python
X = sal_data1[['Age_scaled','Gender_Encoded','Degree_Encoded','Job_Title_Encoded','Experience_years_scaled']]
Y = sal_data1['Salary']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
df = pd.DataFrame({'y_Actual':y_test,'y_Predicted':y_pred})
df['Error'] = df['y_Actual']-df['y_Predicted']
df['abs_error'] = abs(df['Error'])
Mean_abs_Error = df['abs_error'].mean()
print(df)
Mean_abs_Error
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6, edgecolors='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)

plt.title('Predicted vs Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error,mean_absolute_error
acc = r2_score(y_test,y_pred)
print(f'Accuracy of the model = {round(acc,4)*100} %')
```

```python
print(f'Mean Absolute Error = {round(mean_absolute_error(y_test,y_pred),2)}')
```

```python
print(f'Mean Squared Error = {round(mean_squared_error(y_test,y_pred),2)}')
```

```python
print(f'Root mean squared error = {round(mean_squared_error(y_test,y_pred),2)**(0.5)}')
```
